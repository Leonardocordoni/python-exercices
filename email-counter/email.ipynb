```python
import pandas as pd
import numpy as np
```

```python
email = pd.read_json("./emails.json")
```

```python
email.head()
```

```python
remetente = email[['remetente']].value_counts()
```

```python
remetente.to_excel("./count.xlsx")
```

```python
email[['nome', 'email_endereco']] = email['remetente'].str.extract(r'^"?([^"<]+)"?\s*<([^>]+)>')
email['nome'] = email['nome'].str.strip()
email['email_endereco'] = email['email_endereco'].str.strip()


email[['nome', 'email_endereco', 'remetente']].head()
```
